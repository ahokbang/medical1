### 평균 컬럼을 추가해서 grade 컬럼에 A,B,C,D,F 학점을 출력하시오.

```python
# 평균 컬럼을 추가해서 grade 컬럼에 A,B,C,D,F 학점을 출력하시오.

df['평균']=df['합계']/5

def add_grade(ttl):
    if ttl >= 90 : return 'A'
    elif ttl >= 80 : return 'B'
    elif ttl >= 70 : return 'C'
    elif ttl >= 60 : return 'D'
    else : return "F"

df['grade'] = df['평균'].apply(add_grade)
df
```

### SW특기 컬럼을 모두 대문자로 변경, NaN 데이터는 모두 PYTHON으로 출력하시오.

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
### SW특기 컬럼을 모두 대문자로 변경, NaN 데이터는 모두 PYTHON으로 출력하시오.

df['SW특기'] = df['SW특기'].str.upper()
df.fillna('PYTHON', inplace=True) # inplace=True 꼭 하기!! 안하면 적용 안돼!!
df
```

### 함수 사용

```python
### SW특기 컬럼을 모두 대문자로 변경, NaN 데이터는 모두 PYTHON으로 출력하시오.

def upper_change(txt):
    # 대문자로 변경
    if pd.isnull(txt):   # pd.isnull 안하면 error 발생!! 
        return "PYTHON"  
    else : 
        return txt.upper()
```

```python
df['SW특기'] = df['SW특기'].apply(upper_change)
```
