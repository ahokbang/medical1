```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호(-) 표기

x = ['han','kim','hong']
y = [20,40,80]
y2 = [50, 30, 90]
y3 = [80, 90, 95]


# 범례 이름 지정
plt.plot(x,y, label='국어')
plt.plot(x,y2, label='영어')
plt.plot(x,y3, label='수학')
# 범례 표시, 위치 설정
# plt.legend(loc='upper left') 
# ncols=n : 한줄에 범례를 n개씩 출력
plt.legend(loc=1, ncols=3) 
# 상단 타이틀 글자변경, 크기변경
plt.title('선그래프', fontdict={'family':'HYGungSo-Bold', 'size':30})
# x, y축 이름 설정
plt.xlabel('이름', loc='right', color='red') # loc = left, right, center
plt.ylabel('국어', loc='top', color='#d9b3ff') # loc = top, center, bottom
```
