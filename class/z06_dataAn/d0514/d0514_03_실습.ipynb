```python
# chipotle.tsv에서 주문 중 가장 많이 주문한 메뉴 5개를 출력하시오.
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv', sep='\t')
df
```

```python
# df.groupby('item_name').max().sort_values('quantity', ascending=False).head() #mc
df.groupby('item_name').sum().sort_values('quantity', ascending=False).head() #mc
```

```python
# tc
# item_name 별 주문한 개수의 합 출력하시오.
# 주문별 그룹
# quantity의 합을 구해야 함.
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5]
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5]
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5].index
df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5].values

# item_name 별 주문한 횟수가 몇번인지 출력하시오.
df['item_name'].value_counts()[:5]
df['item_name'].value_counts()[:5].index
df['item_name'].value_counts()[:5].values
```

```python
# 그래프
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 10 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호(-) 표기

x = df['item_name'].value_counts()[:5].index
y = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5].values
y2 = df['item_name'].value_counts()[:5].values

plt.figure(dpi=200)
plt.title("[ chipotle에서 주문 중 가장 많이 주문한 메뉴 5 ]")
plt.plot(x,y, label = '주문 수')
plt.plot(x,y2, label = '주문 횟수')
plt.xticks(rotation=90)
plt.legend(loc=1)
plt.xlabel('item_name')
plt.ylabel('quantity')
```
