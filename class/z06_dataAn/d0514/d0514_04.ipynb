```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호(-) 표기
```

```python
x = [1,2,3]
y = [12,9,8]

# 그래프 그리기
plt.plot(x,y, linewidth=5, label='국어', marker='o', ms=20, mec='red', mfc='purple', alpha=0.7, ls=':') 
# linewidth: 선 두께
# label : 범례
# marker : o, x 등 marker 모양
# ms(markersize, ms) : marker 크기
# mec(markeredgecolor, mec) : marker 테두리 색, 
# mfc(markerfacecolor, mfc) : marker 내부 색
# alpha : 투명도
# ls(linestyle, ls) : 선 타입
plt.legend(loc=(0.2, 0.5)) # 위치지점 변경
# plt.legend(loc='lower right') # loc = upper, center, lower
plt.title('꺽은 선 그래프')
plt.yticks([5,7,9,11,13,15]) # y축 눈금 조정
plt.xticks([1,2,3]) # x축 표시 간격 조정==> 눈금 조정
```
