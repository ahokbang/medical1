```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 10 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호(-) 표기

x = [1,2,3]
y = [2,4,8]
```

```python
# 라인스타일
plt.plot(x,y, label='국어', ls='--', marker='o')
plt.legend(loc='upper center')
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 키를 꺽은 선 그래프로 출력하시오.
# 이름 , 키
# y축 표시 : 170, 180, 190, 200, 210

x = df['이름']
y = df['키']

plt.title('키 그래프')
plt.plot(x,y, linewidth=5, label ='키', marker='o', ms=10, mec='blue', mfc='yellow') 
plt.legend(loc='upper center') # 위치지점 변경
for i, txt in enumerate(y):
    plt.text(x[i], y[i]+1, txt, ha='center', color='blue')
plt.ylim(165, 205) # plt.ylim : 축 눈금 범위 지정, ylim 사용하면 글자 크기에 따라 자동으로 범위 표시 되지 않음.
# plt.yticks([150,160, 170, 180, 190, 200, 210]) # y축 눈금 표시
```

```python
# 키를 막대 그래프로 출력하시오.
# 이름 , 키
# y축 표시 : 170, 180, 190, 200, 210

x = df['이름']
y = df['키']

plt.title('키 그래프')
# 가로 막대 그래프
plt.barh(x,y, label ='키') 
# 세로 막대 그래프
# plt.bar(x,y, label ='키') 

plt.legend(loc='upper center') # 위치지점 변경
for i, txt in enumerate(y):
    plt.text(y[i]+5.5, x[i], txt, ha='center', color='blue')

plt.xticks([150,160, 170, 180, 190, 200, 210]) # y축 눈금 표시
# plt.xlim(165, 205)
```
