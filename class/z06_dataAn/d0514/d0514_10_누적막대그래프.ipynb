```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 10 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호(-) 표기
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/score.csv', index_col='지원번호')
df
```

```python
x = df['이름']
y1 = df['국어']
y2 = df['영어']
y3 = df['수학']
```

```python
# 누적 막대 그래프
plt.bar(x,y1, label='국어')
plt.bar(x,y2, bottom=y1, label='영어')
plt.bar(x,y3, bottom=y1+y2, label='수학')

# 값표시
for i, txt in enumerate(y1):
    plt.text(x[i], y1[i]-13, txt, ha='center')
for i, txt in enumerate(y2):
    plt.text(x[i], y1[i]+y2[i]-13, txt, ha='center', color='red')
for i, txt in enumerate(y3):
    plt.text(x[i], y1[i]+y2[i]+y3[i]-9, txt, ha='center', color='blue')

plt.legend(loc='upper center')
```
