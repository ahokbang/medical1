```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 10 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 기호(-) 표기
```

```python
# values = [20, 25, 20, 13, 10, 2] # 값들의 합 : 100
# values = [1,1,1,1,1,1] 
values = [10, 5, 10, 7, 6, 2] 
title = ['python', 'Java', 'Javascript', 'C#', 'C', 'ETC']
explode=[0.1,0,0,0,0,0]
# explode=[0.1]*6 # 다 띄우기

# autopct : 데이터 표시
# startangle : 시작 위치
# counterclock=False : 시계 방향
# plt.pie(values, labels=title, autopct='%.1f%%', startangle=90, counterclock=False) # ==> 90도를 기준으로 시계방향으로 데이터 표시
plt.pie(values, labels=title, autopct='%.1f%%', startangle=90, counterclock=False, explode=explode) 
plt.title("프로그래밍 언어별 선호도")
plt.legend(loc=(1.1,0.3), title='프로그래밍언어')

plt.show()
```
