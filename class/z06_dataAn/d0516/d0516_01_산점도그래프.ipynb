```python
# 학년별 크기 조절
sizes = df['학년']*500 # 1학년: 1*500, 2학년 : 2*500, 3학년: 3*500
# s : 산점도 그래프 점의 크기 조절
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
# 학년별 크기 조절 + 색상 변경
sizes = df['학년']*500 # 1학년: 1*500, 2학년 : 2*500, 3학년: 3*500
# s : 산점도 그래프 점의 크기 조절
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis')
plt.xlabel('영어점수')
plt.ylabel('수학점수')
# ticks : color bar 눈금 표시
# label : 라벨 표시
# shrink : 컬러바 크기 조절
# orientation : 컬러바 위치 하단 배치
plt.colorbar(ticks=[1,2,3], label='학년', shrink=0.3, orientation='horizontal')
```

```python
plt.scatter(df['영어'], df['수학'], label='영수')
plt.scatter(df['영어'], df['과학'], label='영과')
plt.xlabel('영어')
plt.ylabel('수학 및 과학')
plt.legend(loc=1)
```
