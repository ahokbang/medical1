```python
plt.bar(df.columns, df.loc['출생아 수'], color='#ff5c33')
plt.ylabel('출생아 수 (천 명)')
for i, txt in enumerate(df.loc['출생아 수']):
    plt.text(df.columns[i], df.loc['출생아 수'][i]+10, txt, ha='center')
plt.ylim(250,700)
plt.yticks([300,400,500,600])
plt.title("2011~2020년 출생아 수")
```

```python
plt.plot(df.columns, df.loc['합계 출산율'], lw=4 ,marker='o', color='gold', mec="white", ms=10)
plt.ylabel('합계 출산율(가임여성 1명당 명)')
for i, txt in enumerate(df.loc['합계 출산율']):
    plt.text(df.columns[i],df.loc['합계 출산율'][i]+0.1, txt, ha='center')
plt.ylim(0,1.5)
plt.yticks([0,1])
plt.title('2011~2020 출산율')
```

```python
# 그래프 합치기

plt.plot(df.columns, df.loc['출생아 수'])
plt.plot(df.columns, df.loc['합계 출산율'])
```

```python
# fig, ax1 = plt.subplots()
fig, ax1 = plt.subplots(figsize=(10,4))
# suptitle : 상단 타이틀
fig.suptitle('출생아 수 및 합계 출산율')

# 첫번째 막대 그래프
ax1.bar(df.columns, df.loc['출생아 수'], color='#ff812d')
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i, txt in enumerate(df.loc['출생아 수']):
    plt.text(df.columns[i], df.loc['출생아 수'][i]+8, txt, ha='center')


# 두번째 꺽은선 그래프
# 1개의 그래프에 내용을 2개 넣기 ==> 이중 y축 설정
ax2 = ax1.twinx()
# mew(markeredgewidth, mew) : 마커 테두리 두께
ax2.plot(df.columns, df.loc['합계 출산율'], color='#ffd100', lw=4 ,marker='o', mec="white", mew=3 ,ms=13)
ax2.set_ylabel('합계 출산율(가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
for i, txt in enumerate(df.loc['합계 출산율']):
    plt.text(df.columns[i],df.loc['합계 출산율'][i]+0.07, txt, ha='center')

plt.show()
```
