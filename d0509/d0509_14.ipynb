```python
import pandas as pd
# ../ : 폴더를 한칸 나옴(상위폴더)
# ./ : 현재폴더
# / : root 폴더
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 사회 컬럼을 삭제, 음악 컬럼 추가
# 음악 점수는 과학 점수가 80점 이상이면 80점, 80점 미만이면 65점으로 입력
# 합계 컬럼 추가, 국어+영어+수학+과학+음악
# 음악 컬럼의 위치는 과학 컬럼 뒤로 변경

df.drop(columns=['사회'], inplace=True)
df['음악']=65
df.loc[df['과학']>=80,'음악'] = 80
df['합계'] = df['국어']+df['영어']+df['과학']+df['음악']

a_list = list(df.columns) 
df=df[a_list[:6]+[a_list[8]]+[a_list[7]]+[a_list[9]]]

df
```
