```python
# 출생 B:M 2행 제외, 출상애수(명)을 int로 변경

import pandas as pd
df = pd.read_excel('../z20_data/20122022_출생.xlsx', skiprows=2, index_col=0, usecols="B:M")
df.index.name = '목록'
df.head(3)    
df.iloc[0] = df.iloc[0].str.replace(',','').astype(int)

# row의 합 구하기
df.iloc[0].sum()

# 컬럼을 추가
df['합계'] = 0

# 위치 지정해서 데이터를 입력하면 수정
df.iloc[0,-1] = df.iloc[0].sum()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 3번 영어
# 4번 수학
# 5번 국어,영어 를 Nan 처리하시오.

import numpy as np

df.loc['3번','영어'] = np.nan
df.loc['4번','수학'] = np.nan
df.loc['5번',['국어','영어']] = np.nan
```

```python
df
```

```python
# 총합 컬럼을 만들어서, 국어, 영어, 수학, 과학, 사회 합계 점수를 총합에 입력해서 출력하시오.

df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
```

```python
df # NaN값으로 인해 데이터가 정확하지 않아 ==> 데이터 전처리 필요(NaN을 어떻게 처리할지: 0으로 할지, 평균값으로 해야할지)
# NaN을 0으로 할 경우
df.fillna(0, inplace=True)
# 평균값으로 하고 싶으면,
df['국어'].mean() # 얘를 fillna에 넣어줘.
```

```python
df.info()
```

```python
# 국어, 영어, 수학 float으로 되어 다시 int로 형 변환
df['국어'] = df['국어'].astype(int)
```

```python
df.info()
```

```python
df # NaN값이 0으로 변경되어 있음
```

```python
# 다시 총합을 계산해보면,
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
df.describe()
```

```python
df['결과'] =  '불합격'
```

```python
# 총합 점수가 300점이 넘으면 합격, 
df[df['총합']>300]
df.loc[df['총합']>300,'결과'] = '합격'

# 400점 이상인 학생은 우수
df.loc[df['총합']>400,'결과'] = '우수'
df
```
