```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
def grade(avg):
    if avg >= 90 : return 'A'
    elif avg >= 80 : return 'B'
    elif avg >= 70 : return 'C'
    elif avg >= 60 : return 'D'
    else : return 'F'
```

```python
# 총합(국어+영어+수학+과학+사회) 및 평균의 컬럼을 추가하고
# 평점이라는 컬럼을 만들어서, 90점 이상이면 A, 80점 이상이면 B,
# 70점 이상이면 C, 60점 이상이면 D, 60점 미만이면 F 함수를 사용해서 처리하시오.

df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['총합']/5
df['평점'] = 'A'
df['평점'] = df['평균'].apply(grade)

df
```

```python
def result(avg):
    txt = ''
    if avg >= 70 : txt= "합격"
    else : txt = "불합격"
    return txt
```

```python
# 결과 컬럼을 만들어서 평균이 70점 이상인 학생은 '합격' , 아니면 '불합격'

df['결과'] = '불합격'
df['결과'] = df['평균'].apply(result)
df
```

```python
# lambda 함수로 해보기
# lambda(익명함수) : lambda 매개변수
# def(매개변수)
df['결과'] = df['평균'].apply(lambda score:'합격' if score>=70 else "불합격")
df
```

```python
# 결과 2 컬럼 생성, 65점 초과 : 합격, 65점 재시험, 65점 미만 불합격

df['결과2'] = df['평균'].apply(lambda score:'합격' if score>65 else ( "재시험" if score == 65 else "불합격"))
df
```

```python
# 국어+영어 값이 100점을 넘지 않으면 국어 +20 추가해서 출력하시오.

df['국영'] = df['국어']+df['영어']
df['국어'] = df.loc[df['국영']<100, '국어']+20
df
```

```python
def lower_change(lang):
    # notnull : null이 아닌지 확인
    # isnull : null인지 확인
    
    # lang이 null이 아닌지를 묻고, null이 아니면 소문자로
    if pd.notnull(lang): return lang.lower()
    else : return "python"
```

```python
# 1. 'SW특기'에서 각각 단어를 소문자로 변경(함수 사용), NaN은 python으로 입력해서 출력하시오.

df['SW특기'] = df['SW특기'].apply(lower_change)
df
```
