```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score3.csv', index_col=0)
df
```

```python
# 학교 별 국어점수가 가장 높은 점수를 출력하시오.
df.groupby('school')['kor'].max()

df.groupby(['school','grade'])['kor'].max()
df.groupby(['school','grade'])['kor'].mean()['구로고']
```

```python
df.describe()
```

```python
# 남여 수를 출력하시오.
df.groupby('gender').count() # sw의 경우 null값이 있어 앞의 열들과 달라
# 또는
df.groupby('gender').size()
```

```python
# 학교, 성별로 국어, 영어, 수학 점수 평균을 출력하시오.

df.groupby(['school','gender'])[['kor','eng','math']].mean()
df.groupby(['school','gender'])[['kor','eng','math']].mean().sort_values('kor')
```

```python
df.groupby(['school','gender'])[['kor','eng','math']].mean().sort_values('kor', ascending=False)
```

```python
# 디지털고만 보고 싶어
df.groupby(['school','gender'])[['kor','eng','math']].mean().loc['디지털고']
df.groupby(['school','gender'])[['kor','eng','math']].mean().loc['디지털고']['kor'].loc['Female']
# index는 .loc해서 찾아야해
```

```python
# 합계 컬럼 추가해서 학년별 합계의 평균을 출력하시오.
df['total'] = df['kor']+df['eng']+df['math']
df.groupby('grade')['total'].mean()

df.groupby('grade')['total'].min()
df.groupby('grade')['total'].max()
df.groupby('grade')['total'].sum()


# apply 함수 : (def로 만든) 함수 호출
# agg  : 기본 함수 호출
# .agg 함수 사용하면 .mean(), .min(), .max(), .sum() 함수를 묶을 수 있다.
df.groupby('grade')['total'].agg(['mean','min','max','sum'])
```
