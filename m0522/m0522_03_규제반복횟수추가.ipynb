```python
import pandas as pd
df=pd.read_csv('https://bit.ly/fish_csv')
df
```

```python
# 0. 데이터 확인 : 값 중 NaN/0 여부, 타입, 최소값/최대값, 값의 분포, 각 특성별 데이터 최대값 등 확인
# 1. 데이터 전처리 : data, target 분리
fish_target = df['Species'].to_numpy()
fish_input = df.iloc[:,1:].to_numpy()
# 1.1 train/test set 분리
# from sklearn import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42
)
# 1.2 표준화 작업
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 2. 데이터 실습시키기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, train_target)
# 정답률이 0.8067226890756303/0.85로 낮아 
```

```python
# 3. 데이터 예측
predict = lr.predict(test_scaled[:5])
print(predict)
# 기울기, y 절편의 값을 적용시켜서 z 점수 출력하고, softMax 함수를 적용시켜서 
```

```python
# 4. 데이터 정답률 - train/test
print(lr.score(train_scaled, train_target)) # 0.8067226890756303
print(lr.score(test_scaled, test_target)) # 0.85

# 과소적합
```

```python
# 정답률이 낮을 때 낮은 C규제값과 높은 max_iter로 정답률을 높일 수 있다.

# C : 규제, 값이 올가면 규제가 약해지고, 값이 낮으면 규제가 강해짐.
  # 선형회귀에서 규제는 alpha값으로 지정: 값이 높으면 규제강화, 
# max_iter : 반복횟수, 기본은 100
lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_scaled, train_target)
# 수정 후 정답률 0.9327731092436975/0.925로 상승
```

```python
# 3. 데이터 예측
predict = lr.predict(test_scaled[:5])
print(predict)
# 기울기, y 절편의 값을 적용시켜서 z 점수 출력하고, softMax 함수를 적용시켜서 
```

```python
# 4. 데이터 정답률 - train/test
print(lr.score(train_scaled, train_target)) # 0.9327731092436975
print(lr.score(test_scaled, test_target)) # 0.925
```
