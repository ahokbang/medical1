```python
data = [length, weight, results]
df = pd.DataFrame(data)
df=df.T
# df
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    df.iloc[:,:2], df.iloc[:,-1:], random_state=42
)
# train_input, test_input, train_target, test_target = train_test_split(
#     data, target, random_state=42
# )
```

```python
# 표준화작업
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
# 길이 30, 무게 600
# 길이 9.8, 무게 6.7
new = ss.transform([[25,150]])
new2 = ss.transform([[30,600]])
new3 = ss.transform([[9.8,6.7]])
```

```python
# 데이터 학습시키기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, train_target)
```

```python
# 예측하기
predict = lr.predict(test_scaled[:5])
predict
```

```python
# 정답률
print("train : ", lr.score(train_scaled, train_target))
print("test : ", lr.score(test_scaled, test_target))
```
