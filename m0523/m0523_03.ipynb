```python
# 2. 데이터 전처리
# 2.1 data, result 분리
data = df.iloc[:, 2:-2].to_numpy()
target = df['연봉(2018)'].to_numpy()
```

```python
# 2. 데이터 전처리
# 2.2. train, test 세트 분리

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    data, target, random_state=42
)
```

```python
# 3. 데이터 실습시키기
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_input, train_target)
```

```python
# 4. 데이터 예측
predict = lr.predict(train_input)
predict
```

```python
# 5. 데이터 예측률 구하기
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))

# 0.7253752944192529
# 0.2756747829495232
```

```python
# 특성화를 통한 예측률 높이기
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3, include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
    
# degrees = [2, 3, 4, 5, 6]
# train_polys = []
# test_polys = []

# for degree in degrees :
#     poly = PolynomialFeatures(degree=degree, include_bias=False)
#     poly.fit(train_input)
#     train_polys.append(poly.transform(train_input))
#     test_polys.append(poly.transform(test_input))
```

```python
# 재학습 시키기
lr = LinearRegression()
lr.fit(train_poly, train_target)
```

```python
# 예측률
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))

# 1.0
# -30.13422667598011
```

```python
# 규제 적용하기 ==> 표준화하기
# 표준화

from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
# 릿지규제 학습시키기
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
```

```python
# 정답률
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```

```python
# 릿지규제 알파 추가

train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]

for alpha in alpha_list :
    ridge = Ridge(alpha=alpha)
    ridge.fit(train_scaled, train_target)
    train_score.append(ridge.score(train_scaled, train_target))
    test_score.append(ridge.score(test_scaled, test_target))
    
```

```python
x = np.arange(1,9).reshape(-1,1)
plt.plot(x,train_score)
plt.plot(x,test_score)
```

```python
ridge = Ridge(alpha=1000)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```
